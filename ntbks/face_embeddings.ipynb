```python
def grab_name(file_name):
    names = re.split("[_\.]", file_name)[-3:-1]
    return f"{names[0]}_{names[1]}"
grab_name(imagePath)
```

```python
# grab the paths to the input images in our dataset
cand_fold = "./data/candidate_images/*.jpg"
imagePaths = glob.glob(cand_fold)
print(imagePaths[:5])
```

```python
def encode_face(imagePath):
    '''Encode 1 face in an input image.'''
    face_encoding = None
    
    image = cv2.imread(imagePath)
    if image is not None:
        rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # dlib ordering (RGB)
        # detect the (x, y)-coordinates of the bounding boxes of each face
        boxes = face_recognition.face_locations(rgb, model="hog") # cnn or hog

        # compute the facial embedding for the face
        encodings = face_recognition.face_encodings(rgb, boxes)
        if len(encodings) > 0:      
            face_encoding = encodings[0].tolist()
            
    return face_encoding
```

```python
# loop over the image paths
nimages = len(imagePaths)
candidate_face_encodings = {}
for (i, imagePath) in enumerate(imagePaths):
    print(f"[INFO] processing image {i}/{nimages}")
    name = grab_name(imagePath)
    face_encoding = encode_face(imagePath)
    
    candidate_face_encodings[name] = {
        "image_path": imagePath,
        "face_encoding": face_encoding
    }
    
    
```
