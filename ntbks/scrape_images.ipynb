```python
import pandas as pd
import requests

from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
##-------------
## Submit Query to GDELT
##-------------
url = u'https://en.wikipedia.org/wiki/2018_United_States_Senate_elections'

html = requests.get(url).text
soup = BeautifulSoup(html, 'html.parser')

state_sums = soup.select('.infobox')


rez = {}
for state_sum in state_sums:
    state_name = state_sum.find('caption').text
    
    box = state_sum.find_all('tr')[3]
    potential_links = box.find_all('a')
    for pl in potential_links[2:-2]:
        link = pl['href']
        try:
            title = pl['title']
        except:
            title = None
        text = pl.text
        if len(text.split(" ")) in range(2,4):
#             print('---', link, "/", title, "/", text)
            print((text, f"https://en.wikipedia.org{link}"))
            rez[state_name] = rez.get(state_name, []) + [(text, f"https://en.wikipedia.org{link}")]
            
```

```python
pd.DataFrame.from_dict(rez, orient="index")
```
